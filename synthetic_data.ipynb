```python
import cv2

def extract_frames(video_path, output_folder, n):
    # Open the video file
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error: Could not open video file {video_path}")
        return
    
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    
    print(f"Total frames: {frame_count}")
    print(f"FPS: {fps}")

    current_frame = 0
    saved_frame_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        if current_frame % n == 0:
            frame_filename = f"{output_folder}/frame_{saved_frame_count:05d}.jpg"
            cv2.imwrite(frame_filename, frame)
            saved_frame_count += 1
            print(f"Saved frame {saved_frame_count} at position {current_frame}")

        current_frame += 1

    cap.release()
    print("Finished extracting frames.")

# Example usage
video_path = "D:\\core9_work\\Amir's Land rover AI Enablement\\1.mp4"  # Path to the input video
output_folder = "D:\\core9_work\\Amir's Land rover AI Enablement\\dummy_chromadb"  # Folder where extracted frames will be saved
n = 150  # Extract every n-th frame

extract_frames(video_path, output_folder, n)
```

```python
import cv2
import os
from datetime import datetime
from datetime import datetime, timedelta


def extract_frames(video_path, output_folder, timestamps_file):
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Open video file
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error: Could not open video.")
        return
    
    fps = cap.get(cv2.CAP_PROP_FPS)  # Get frames per second
    interval = int(5 * 30)  # Interval in frames (5 seconds)
    counter=206
    frame_count = 0
    custom_date = datetime(year=2024, month=6, day=20, hour=15, minute=5, second=15)
    with open(timestamps_file, 'w') as f:
        while True:
            ret, frame = cap.read()
            if not ret:
                break
    
            if int(frame_count % interval )== int(0):
                print(frame_count,"---->>",frame_count % interval)
                #date_time_str = datetime.strftime("%d-%m-%Y %H:%M:%S")
    
                custom_date = custom_date + timedelta(seconds=5)
                img_filename = os.path.join(output_folder, f"frame_"+str(counter)+".jpg")
                counter=counter+1
                cv2.imwrite(img_filename, frame)  # Save the frame as an image
                f.write(f"{custom_date}\n")
    
            frame_count += 1
    
        cap.release()

        
    print(f"Frames extracted and saved in '{output_folder}'. Timestamps saved in '{timestamps_file}'.")
    print(frame_count)


# Usage example
video_path = "D:\\core9_work\\Amir's Land rover AI Enablement\\4.mp4"  # Path to the input video
output_folder = "D:\\core9_work\\Amir's Land rover AI Enablement\\dummy_chromadb"
timestamps_file = "D:\\core9_work\\Amir's Land rover AI Enablement\\timestamps3.txt"

extract_frames(video_path, output_folder, timestamps_file)
```
